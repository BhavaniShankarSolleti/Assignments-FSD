**3. Define acessibility in the context of web development. Discuss why its essential to create acessible websites and how it benefits differnet user groups ?**

**How it helps Different User Groups**

**For Visual Impaired people**: Screen readers, high-contrast modes, and braille displays help users navigate digital content.

**For people with hearing impairments**: Captions, transcripts, and visual alerts make information accessible.

**Motor Disabilities**: Voice commands, adaptive keyboards, and switch controls assist in interaction.

**Cognitive Disabilities**: Simple layouts, clear navigation, and easy-to-understand content improve usability.

**Elderly Users**

--> Larger text sizes, voice assistance, and simplified interfaces make digital tools easier to use.

**Temporary Disabilities**

--> Individuals recovering from injuries (e.g., a broken arm) can benefit from voice controls and alternative input methods.

**Situational Limitations**

--> People in noisy environments will rely on captions.

--> Those in bright sunlight benefit from high-contrast designs.

**General Public**

--> Accessibility features like captions and voice assistants improve convenience for everyone.

--> Clear design and intuitive navigation benefit all users, regardless of ability.
