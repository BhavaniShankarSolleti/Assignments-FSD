Task 2:


Problem Statement

Explain the difference between Absolute and Relative positioning.

Ans:

Absolute positioning means:
* The element with property position and value absolute can have proerties like top, bottom, left, right.
* The position of the element depends upon the nearest positioned ancestor with position value other than the static.
* The element is removed from the regular document flow.

Relative posistioning means:
* The element with property position and value relative can also have the properties like top, bottom, left, right.
* Position of element can be adjusted with respect to its original position position.
* The element is not removed from the regular document flow.

Problem Statement

Explain the z-index, with a code example.

Ans:

  **Z-Index**
* Z-Index in css is used to control the stacking order of elements in z-axis(the axis that comes out of the screen to the user).
* Elements with higher z-index will present above the elements with lower z-index and vice-versa.
* This only works, if this element presents inside another positioned element otherwise not.
* Ex:

  //css

  .first
  {
  z-index:3;
  }

  .second{
  z-index:2;
  }

  .third{
  z-index:1
  }

* In the above example, the first element is at bottom, second one is in the middle and third one is ar top.
* Possible values are positive, negative , zero.
