1. Explain the purpose, and importance of the OSI Model, and briefly describe each layer of the OSI model
with an example.

Ans:

**OSI model**

* OSI(open systems interconnection) model is a conceptula framework which standarzises how different networking systems communicate over a netwrok

**Purpose of OSI model**
* Standardization: Provides a universal standard for different systems to communicate.
* Interoperability: Ensures devices from different vendors can work together.
Modularity: Break comminication into different layers. so that each layer can be designed/updated independently.
* Troubleshooting: Helps diagnosing networking problems by checking layer by layer.
* Learning and teaching: Makes learning and teaching complex networking easier to understand.

**Importance of OSI model**
* It provides a blueprint for networking.

There are 7 layers in the OSI model

1. Physical layer
* Deals with 0s and 1s over physical medium.
* Concerned with cables, switches, hubs, voltages, frequencies, signal.
* Ethernet cables, fiber optics, Wifi signals.
* function: Transmission of raw data.

2. Data-link layer
* Ensures error free data transfer between nodes.
* Responisble for MAC addresses, frames, error detection (crc).
* Sub layers
  * LLC (logical link control)
  * MAC(Media access control)
* Ex: Ethernet, switches, ppp.
* Function: Framing, error detection, flow-control.

3. Network layer
* Responsible for routind data (choosing best path).
* Uses IP addresses for logical addressing.
* Ex: Routers, IPv4, Ipv6.
* Function: packet forwarding, routing, addressing.

4. Transport layer
* Provides end to end communication between devices.
* Ensure reliable delivery (or fast but unreliable).
* Protocols:
  * TCP - Reliable, error-checked(ex:file-tranfer, web-browsing).
  * UDP - Fast, No guarantee(ex: video streaming, gaming).
Function: Segmentation, error recovery, flow control.

5. Session layer
* Manages sessions(connections) between applications.
* Responsible for opening, maintaining, and closing sessions
* Ex: Remote procedure calls, authetication.
* Funtion: Session management, synchronization.

6. Presentation layer
* Deals with data translation, encryption, compression.
* Ensures data is in a readable format for the application.
* Ex: SSL/TLS (encryption), JPEG, MP3, ASCII, EBCDIC.
* Function: Data formatting and security.

7. Application Layer (Layer 7)
* Closest to the end-user.
* Provides network services directly to applications.
* Ex: HTTP, FTP, SMTP, DNS, POP3.
* Function: Application services (web, email, file transfer).











2.Explain the purpose, and importance of the TCP/IP, and briefly describe the different layers of the TCP/IP model.


Ans:

**TCP/IP (Transmission Control Protocol / Internet Protocol)**

It's a practical networking model that describes how data is transmitted across the internet. It is the foundation of the modern internet.

**Purpose of TCP/IP Model**

* Practical Implementation → Unlike OSI, TCP/IP is actually used in real systems.

* Standardization → Defines a universal protocol suite for communication.

* Interoperability → Allows different networks/devices to communicate globally.

* Flexibility → Works across LAN, WAN, and the entire Internet.

* Scalability → Supports billions of devices worldwide.

**There are 4 Layers in the TCP/IP Model**

1. **Network Access Layer (a.k.a. Link Layer)**

* Corresponds to OSI's Physical + Data Link Layers.

* Handles hardware transmission of data over the physical medium.

* Deals with MAC addresses, Ethernet, Wi-Fi, device drivers.

* Protocols/Technologies: Ethernet, Wi-Fi, ARP, PPP.

* Function: Defines how data is physically sent over a network.

2. **Internet Layer**

* Corresponds to OSI's Network Layer.

* Provides logical addressing and routing.

* Each device has an IP address.

* Protocols: IPv4, IPv6, ICMP (ping), IPsec.

* Function: Routing packets across multiple networks.

3. **Transport Layer**

* Corresponds to OSI's Transport Layer.

* Ensures end-to-end communication between applications.

* Protocols:

   * TCP (Transmission Control Protocol) → Reliable, connection-oriented (web, email, file transfer).

   * UDP (User Datagram Protocol) → Fast, connectionless (video streaming, gaming, VoIP).

* Function: Error checking, sequencing, flow control.

4. **Application Layer**

* Corresponds to OSI's Application + Presentation + Session Layers combined.

* Closest to the end-user applications.

* Provides services like web browsing, email, file transfer.

* Protocols: HTTP, HTTPS, FTP, SMTP, POP3, DNS, DHCP, Telnet, SSH.

* Function: Enables user interaction with network services.

4. What is DNS, and how does it resolve domain names?

Ans:

**DNS**
* DNS stands for domain name system.
* Its like of "phonebook of the internet of ip addresses".
* Translates  domain names to thier corresponding IP addresses like (142.150.234.23).
* Without this we should remember long IP addresses for every webiste.

**How DNS resolves domain names to up addresses**

* Suppose let's assume www.google.com  

1. **Browser Cache Check**

* Your browser first checks its cache.

* If it has already visited www.example.com recently, it retrieves the stored IP address directly.

2. **Operating System Cache**

* If not in the browser cache, the OS checks the local DNS cache.

3. **Router/ISP DNS Resolver**

* If still not found, the request goes to your ISP's DNS resolver (a recursive DNS server).

4. **Root DNS Server**

* If the ISP's resolver doesn't know, it queries a Root DNS Server.
* The root server doesn't know the exact IP, but it knows where to find info about Top-Level Domains (TLDs) (like .com, .org, .net).

5. **TLD DNS Server**

* The resolver then asks the TLD server (for .com in this case).

* The TLD server points to the Authoritative DNS Server for example.com.

6. **Authoritative DNS Server**

* The authoritative server contains the actual DNS records (like A, AAAA, MX, etc.).

* It returns the IP address of www.google.com (say 93.184.216.34).

7. **Response Back**

* The resolver sends the IP back to your browser.

* The browser can now connect directly to the web server at that IP.


**Types of DNS records**

* A Record → Maps a domain to an IPv4 address.
* AAAA Record → Maps a domain to an IPv6 address.

* CNAME → Alias for another domain (e.g., www → example.com).
* MX Record → Mail exchange servers for email.
* NS Record → Specifies authoritative name servers for the domain.

5.Provide a brief description of the HTTP request and response model along with an illustrative diagram.


Ans:
The HTTP (hyper text tansfer protocol) request and response model is the foudation for communication between client and server over the internet.It follows a request response model where the client sends the request to the server and server responds with the HTTP response.

**HTTP request**

*  Initiated by the client to request resources or perform actions on the server.
*  Components include
    
    * Request line: Specifies the HTTP method  (GET, POST, PUT, DELETE), the resourse url and the http version.
    * Headers: Provide metadata (Ex: conent type, authentication tokens etc)
    * Body(optional):Contains data sent to the server (Ex: form data in post requests.)
* A browser sends a GET request to retrieve a webpage.

**HTTP response**
* HTTP response sent by the server to the client in response to its request.
* Components include:
  * Status line: Contains HTTP version, status code(ex: 200 ok, 404 not found) and reason phrase.
  * Headers: Metadata about the response (Ex: content type, server details).
  * Body(optional): Contains the requested resource (ex:html, json, images ).
* Ex: The server returns an html page or error message.
* The overall process is statless, meaning each request-response pair is independent unless managed by cookies or sessions.

6.What is CORS? Explain its purpose and why developers frequently encounter CORS errors during
development.


Ans:

**CORS (Cross origin resource sharing)**

* It is a security mechanism implemented by web browsers to control how resources on one domain(Ex: API's, images scripts) can be requested by different domain.
* An origin is defined as the combination of protocol,domain and port.Ex:
http://api.example.com and http://api.example.com are different origins.
* cors is enforced by browsers through the same origin policy, which restricts web pages from making requests to a different origin unless explicitly allowed by the server. cors provide a way to servers to specify which origins, methods, and headers are permitted in cross -origin  requests using http headers.

**Purpose of CORS**

* The primary purpose of cors is to enhance web security by preventing unauthorised cross-origin requests that coudld lead to malicious acticities , such as unauthorised access, malicious scripts etc.
* Preventing unauthorised access: Ensures that only trusted origins can access sensitive resources (ex: api's containing user data).
* protecting against malicious scripts: Blocks scripts running on website from making unauthorised requests to another site (Ex: to steal user data or perform action without consent).
* Enabling controlled sharing: Allows servers to explicitly permit cross origin requests from specific domains, or headers, balancing security and functionality.

* For example, a web app at https://frontend.com can safely request data from an api at https://api.backend.com if the server allows it via cors headers.

**Issues often encounter by developers and the solution for each issue**

1. Server not configured for CORS:
   * The server does not include the necessary headers(Ex: Access-Control-Allow-Origin) in its response.
   * Ex: A frontedn at http://localhost:3000 tries to fetch data from https://api.example.com, but the api server does not allow the origin local host.
   * Fix: The server must be configured to include Access-control-origin:http://localhost:3000 or Access-control-origin :* for public api's through less secure.
2. Mismatch in origins:
   * The request originates from differnet protocol,domain, or port than what the server expects.for example, http://example.com and http://api.example.com are different origins.
   * Ex: A developer testing a frontend app localy in his machine but in server in api he confgured with different origin this lead to error, developer cant test the api.
   * Fix: Update the server to allow specific origin for testing purposes and reconfigured after the testing is over or use proxy development( a local development server that forward requets
3. preflight request issues:
   * For non-simple requests(Ex: post with cutom headers or content types like application/json). The browser sends a prefligh options request to check if the server allows the request or not. If the server doesn't respond correctly(Ex: missing Access-control-allow-methods or access-control-allow-headers) the browser blocks the request.
   * Fix: Ensure the server handles the options requests and returns the appropriate cors headers.
4. misconfigured or overly restrictive cors policies
   * The server may allow only specific origins, methods, or headers and the clients request doesnot match with these criteria.
   * Ex: The server allows only get request, but the server allows only post request.
   * Fix: Adjust the cors policy to allow the the required methods, headers or origins.
5. Third party api restrictions
   * when uisng third party apis, developers may encounter cors errors if the api doesn't allow request from thier frontend origin(Ex: a browser base app).
   * Fix: use server side proxy to make the api call or check if the api provider offers a cors enabled endpoint.
6. authentication and credentials:
   * If a request includes credentials (EX: cookeis or tokens, the server must explicitly allow this with Access-control-Allow-credentials:true and specify an exact origin not(*).If misconfigured cors error occur.
   * Fix:Ensure the server correctly handles credentield request and includes the appropraite headers.
