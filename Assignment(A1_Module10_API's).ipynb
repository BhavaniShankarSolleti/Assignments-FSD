2.Explain the core principles of REST APi's ?
Ans:
1. Client-server architecture
   * The client( Ex: react frontend) and the server (Ex: Node.js/express backend) are separate.
   * Client requests -> server responds -> clear seperation of responsibilites.
   * This makes apps more scalable and flexible.

2. Statelessness
   * Each request from client to server must contain all necessary information.
   * Server doesn't store client context between requests.
   * Ex: Client sends login credentials -> Server verifies -> server gives a token.
   * In future request's, the client must send that token again.

3. Uniorm interface
   * A consistent of interacting with resources using standard rules.
   * Resources identified by url's
   * Ex: /users/101 refers to user with id 101.
   * Standard HTTP methods
      * Get -> Retrive
      * POST -> Create
      * PUT/PATCH -> Update
      * DELETE -> Remove
   * Consistent strucure: JSON is common for request/response format.

4. Resource based
   * Everthing in REST is treated as resource(users, order, products,payments).
   * Ex: /product/45 -> product with id 45
5. Layered system
   * A REST API may have multiple layers (load balancers, authentication, servers, caching).
   * Client does not need to know whether it's directly talking to the backend or through intermediaries.
6. Cacheability
   * Responses should indicate whether they can be cached( to improve performance).
   * Ex: A list of products might be cached, but user login details should not.
7. Code on demand
   * server can provide executable code to the client to extend functionality.
   * Not commonly used, but part of the REST principles.

Summary: RESTful API's are resource based, stateless, cacheable, uses standard http methods and ensure uniform layered architecture.


3.What are the common HTTP methods used in REST APIs and what are their typical uses?

Ans:

**1.GET**
  *  Purpose: Retrieve a resource.
  *  safe and idempotent: Doesn't change data, can be called multiple times with the same result.
  *  Ex:
     * GET/users -> Get all users
     * GET/users/101 -> Get user with ID 101

**2.POST**
  * Purpose: Create a new resource.
  * Not idempotent: Multiple calls create multiple resources.
  * Ex: POST/users with body {name:"shankar"} - > creates new user named alice.


**3.PUT**

   *  Purpose: Update or replace an entire resource.

   * Idempotent: Multiple calls with the same data won’t change the result after the first update.

   * Ex: PUT /users/101 with body { "name": "Alice", "age": 25 } → Replaces user 101’s details with the new data


**4.PATCH**

  * Purpose: Partially update a resource.

  * Idempotent (usually): Updates only specified fields.

  * Ex: PATCH /users/101 with body { "age": 26 } → Updates only the age of user 101

**5.DELETE**

   * Purpose: Remove a resource.

   * Idempotent: Deleting the same resource multiple times has the same effect (resource gone).

   * Ex: DELETE /users/101 → Deletes user with ID 101


**6.HEAD (less common but useful)**

   * Purpose: Same as GET but returns only headers, not the body.

   * Typical Use: To check if a resource exists or get metadata.

   * Ex: HEAD /users/101 → Checks if user 101 exists


**7.OPTIONS (used for discovery & CORS)**

 * Purpose: Returns the allowed HTTP methods for a resource.

 * Typical Use: Pre-flight requests in CORS (Cross-Origin Resource Sharing).

 * Ex: OPTIONS /users → Might return GET, POST, PUT, DELETE


4.Explain the difference between stateless and stateful REST API's ?

Ans:

**Stateless vs Stateful REST APIs**

**1.Stateless REST APIs**

* Definition: The server does not store any information about the client’s session.

* Every request from the client must include all necessary information (authentication, parameters, body).

* The server treats each request as independent.


**Advantages:**

* Scalable (easy to add more servers behind load balancers).

* Reliable (server crash doesn’t lose session info).

* Easier to cache responses.

**Disadvantages:**

* Client must send authentication (like token) every time.

* Slightly more network overhead since request is self-contained.


* Example:

  * A shopping site REST API where each request includes the user’s authentication token and cart info.

  * GET /cart?user_id=101&token=abc123 → Server processes and responds, without remembering previous requests.


**2.Stateful REST APIs**

* Definition: The server remembers client state between requests.

* The client and server maintain a session (using session IDs, cookies, etc.).

* Each new request depends on the context of previous interactions.


**Advantages:**

* Less data needs to be sent in every request (server remembers session).

* Simpler for clients, since the server tracks their state.

**Disadvantages:**

* Harder to scale (server must keep track of sessions).

* If a server crashes, session info may be lost.

* Caching becomes difficult.


* Example:

  * A banking REST API where login creates a session, and subsequent requests rely on that session ID.

  * Client logs in once → server keeps session in memory → next request doesn’t need to resend full auth details.

* Most REST APIs are stateless (to keep them scalable and simple).

* Stateful APIs are used only when the business logic requires maintaining session state (e.g., transactions).
