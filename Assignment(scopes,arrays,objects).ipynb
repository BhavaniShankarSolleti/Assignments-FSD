    => const shoppingCart = ['Milk', 'Coffee', 'Tea', 'Honey']

*    remove 'Honey' if you are allergic to honey
*    modify Tea to 'Green Tea'



```python
function isPresent(arr,item){
    for(let i =0;i<arr.length; i++){
        if(arr[i]=='item'){
            return true;
        }
    }

}
function addItem(isPresent,item,arr){
    if(isPresent=='true' || item=='Meat'){
       arr.unshift(item);
    }else if(isPresent=='true' || item=='Sugar'){
        arr.push(item);
    }else{
        return;
    }

}

function replaceItem(arr,item1,item2){
     for(let i=0; i<arr.length;i++){
        if(arr[i]==item1){
            arr[i]=item2;
        }
     }
}

function removeItem(arr,item){
    for(let i=0;i<arr.length;i++){
        if(arr[i]==item){
            arr.splice(i,1);
        }
    }
}

const shoppingCart = ['Milk', 'Coffee', 'Tea', 'Honey'];
addItem(isPresent(shoppingCart,'Meat'),'Meat',shoppingCart);
addItem(isPresent(shoppingCart,'Sugar'),'Sugar',shoppingCart);
removeItem(shoppingCart,'Honey');
replaceItem(shoppingCart,'Tea','Green Tea');
console.log(shoppingCart);
```

```python
function isExtensible(object){
       let lenBeforePropAdd=Object.keys(object).length;
        object.property="value";
        let lenAfterPropAdd=Object.keys(object).length;
        if(lenBeforePropAdd!=lenAfterPropAdd){
            return true
        }else{
            return false;
        }
}

function isSealed(object){
        let keyArray = Object.keys(object);
        if(!isExtensible(object)){
            let lenBeforePropDelete =keyArray.length;
            let property = keyArray[0];
            delete object[property];
            let lenAfterPropDelete = keyArray.length;
            if(lenBeforePropDelete==lenAfterPropDelete){
                return true;
            }else{
                return false;
            }
        }

        return false;
}

const teacher = {
         subject:'Math'
}

const student ={
    name:"bhavani",
    age:24,
    place:"dharmavaram",
    college:"sjcit",
    school:'sadhana EM high school'
}
Object.preventExtensions(student);
let entensibleStatus = isExtensible(student);
console.log(isExtensible(student));
Object.seal(teacher);
let sealedStatus = isSealed(teacher);
console.log(sealedStatus);
```

```python

// a. Add a Student
function addStudent(studentsArray, newStudent) {
  if (studentsArray.some(student => student.id === newStudent.id)) {
    console.log(`Student with ID ${newStudent.id} already exists.`);
    return;
  }
  studentsArray.push(newStudent);
  console.log("Student added successfully.");
}

// b. Update Student Information
function updateStudentInfo(studentArray,studentID,updateInfo){
        const index = studentArray.findIndex(student=>student.id===studentID);
        if(index!=-1){
        for(let key in updateInfo){
            studentArray[index][key]=updateInfo[key];
            }
        }else{
            console.log(`student with ${studentID} not found`);
        }
    }

// c. Delete a Student
  function deleteStudent(studentsArray, studentId) {
  const initialLength = studentsArray.length;
  const filteredStudents = studentsArray.filter(student => student.id !== studentId);
  if (filteredStudents.length < initialLength) {
    // Update the original array reference (or return the new array)
    studentsArray.length = 0; // Clear the original array
    studentsArray.push(...filteredStudents); // Push elements from the filtered array
    console.log(`Student with ID ${studentId} deleted successfully.`);
  } else {
    console.log(`Student with ID ${studentId} not found.`);
  }
}

// d. List All Students
function listAllStudents(studentsArray) {
  if (studentsArray.length === 0) {
    console.log("No students in the system.");
    return;
  }
  console.log("List of All Students:");
  studentsArray.forEach(student => {
    console.log(`ID: ${student.id}, Name: ${student.firstName} ${student.lastName}, Age: ${student.age}, Grade: ${student.grade}`);
  });
}

// e. Find Students by Grade
function findStudentsByGrade(studentsArray, grade) {
  const studentsWithGrade = studentsArray.filter(student => student.grade === grade);
  if (studentsWithGrade.length === 0) {
    console.log(`No students found with grade "${grade}".`);
    return [];
  }
  console.log(`Students with grade "${grade}":`);
  studentsWithGrade.forEach(student => {
    console.log(`ID: ${student.id}, Name: ${student.firstName} ${student.lastName}`);
  });
  return studentsWithGrade;
}

// f. Calculate Average Age
function calculateAverageAge(studentsArray) {
  if (studentsArray.length === 0) {
    console.log("No students to calculate average age.");
    return 0;
  }
  const totalAge = studentsArray.reduce((sum, student) => sum + student.age, 0);
  const average = totalAge / studentsArray.length;
  console.log(`Average age of students: ${average.toFixed(2)}`);
  return average;
}
```
